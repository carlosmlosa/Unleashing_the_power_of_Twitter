```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
#Clustered tweets
# df = pd.read_csv('/kaggle/input/us-senate-analysis/dataset_with_topics.csv')
df = pd.read_csv("/content/dataset_with_topics.csv")
df.head()
```

```python
# Topic list manually made
# topics = pd.read_csv('/kaggle/input/us-senate-analysis/topic_info_150.csv').iloc[:,0:4]
topics = pd.read_csv("/content/topic_info_150.csv").iloc[:,0:4]
topics.head()
```

```python
# Senators info
# senators = pd.read_csv('/kaggle/input/us-senate-analysis/us_senators.csv')
senators = pd.read_csv('/content/us_senators.csv')
senators.loc[senators["Party"].isin(['Independent[d]','Independent[n]', 'Democratic (DFL)[p]']), "Party"] = "Democratic"
senators.head()
```

Merge df and topics to get the topics

```python
df = df.merge(topics, left_on='topic_150', right_on='Topic')
df.head()
```

```python
# Clean
df.drop(['topic_1500', 'topic_500', 'Topic', 'Name', 'id'], axis=1, inplace =True)
```

```python
# reduce senators dataset
senators = senators[['Party', 'screen_name', 'Senator',"followers_count","friends_count","x","y","State", "name"]]
df = df.merge(senators, left_on='username', right_on='screen_name')
df.head()
```

```python
sen_pop = df.groupby("Senator").count()
sen_pop = sen_pop[["text","retweets","likes","name"]]


# Get sum of rt and likes from all the tweets
sen_pop["retweets"] = df.groupby("Senator")["retweets"].sum()
sen_pop["likes"] = df.groupby("Senator")["likes"].sum()


# Include Senators information
sen_pop = sen_pop.merge(senators, on='Senator', how='right')
sen_pop.set_index('Senator', inplace=True)
```

## Popularity equation
Our own popularity expression is based on the metrics that are more correlated. It is undeniable that the number of likes and retweets accurately represents the number of interactions; however, to maintain the proportion and to penalize having many tweets with few interactions, we decided to divide by the total number of tweets.

We have also observed a significant correlation between followers count and the number of interactions with the tweet. In order to take into consideration for that metric as well, we decided to add a factor the equation.

$ popularity = \frac{rt+likes+0.25\cdot followers}{tweets}$

# Retweets and likes plots

```python
# Compute popularity
sen_pop["popularity"] = ((sen_pop["retweets"]+sen_pop["retweets"]+0.25*sen_pop["followers_count"])/sen_pop["text"])
sen_pop = sen_pop.sort_values(by='popularity', ascending=False)


# Normalization with respect to the max
sen_pop["popularity"] = sen_pop["popularity"]/sen_pop["popularity"].max()
sen_pop.head()

most_pop = sen_pop.iloc[0]
all_sen_pop= sen_pop
```

The popularity of each tweet is calculated using the expression $ popularity = rt+likes$

```python
def plot_line_Tweet_activity(x, title):
    retweets_mean = x['retweets'].mean()
    likes_mean = x['likes'].mean()
    # print("Average #Retweets:", retweets_mean)
    # print("Average #Likes:", likes_mean)

    fig = px.line(x, x='date', y=[ 'retweets', 'likes'],
                  labels={
                     "date": "Time",
                    #  "text": "Number of tweets",
                     "retweets": "Number of retweets",
                     "likes": "Number of likes"
                  })

    fig.update_layout(title_text=title)

   # Add horizontal dotted lines for mean values of retweets and likes
    fig.add_trace(go.Scatter(x=[x['date'].min(), x['date'].max()], y=[retweets_mean, retweets_mean],
                             mode='lines', name=f'Avg #Retweets ({retweets_mean:.2f})', line=dict(color='blue', dash='dot')))
    fig.add_trace(go.Scatter(x=[x['date'].min(), x['date'].max()], y=[likes_mean, likes_mean],
                             mode='lines', name=f'Avg #Likes ({likes_mean:.2f})', line=dict(color='red', dash='dot')))


    fig.show()

def print_tweets(df):
  for index, row in df.iterrows():
    print("\nTweet: ",row.text)
    print("Date: ",row.date)
    print("Likes: ", row.likes)
    print("Rts: ", row.retweets)
    print("Popularity: ",row.popularity)


def analyze_senator_tweets_popularity(name):
    # Calculate Popularity of the tweets
    df["popularity"] = df["likes"]+df["retweets"]
    df["popularity"] = df["popularity"]/df["popularity"].max()

    # Dataframe with the senator tweets
    username = senators.at[senators.loc[senators.Senator == name].index[0], "screen_name"]
    tweets = df[df['username'] == username].copy()
    print(f"{name}'s tweets average Popularity: ",tweets["popularity"].mean())
    tweets['date'] = pd.to_datetime(tweets['date']) # Convert date column to datetime format
    tweets.sort_values("date")

    # Tweets aggrupated by month
    tweets_month = tweets.groupby([pd.Grouper(key='date', freq='M')]).agg({'text': 'count', 'retweets': 'sum', 'likes': 'sum'}).reset_index()
    plot_line_Tweet_activity(tweets_month, f'{name} Tweet Activity Over Time')

    # Tweets aggrupated by day
    tweets_day = tweets.groupby([pd.Grouper(key='date', freq='D')]).agg({'text': 'count', 'retweets': 'sum', 'likes': 'sum'}).reset_index()
    plot_line_Tweet_activity(tweets_day, f'{name} Tweet Activity Over Time')

    tweets_pop = tweets.sort_values("date")
    fig = px.line(tweets_pop, x='date', y=["popularity"],
              labels={
                  "date": "Time",
                  "popularity": "Popularity"
              })
    fig.update_layout(title_text=f"{name} Tweets Popularity Over Time")
    fig.show()

    # Print most popular tweets
    print_tweets(tweets.sort_values("popularity",ascending = False).head())



analyze_senator_tweets_popularity("Jon Ossoff")
```
